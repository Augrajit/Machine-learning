```python
import pandas as pd
import numpy as np
```

```python
ml_filepath ="/content/sample_data/melb_data.csv"
mlb_data =pd.read_csv(ml_filepath)
```

```python
mlb_data.info()
```

```python
mlb_data.describe()
```

```python
mlb_data
```

```python
mlb_data.hist(bins=50,figsize=(20,15))
```

```python
feat=['Rooms','Bathroom','Landsize','Lattitude','Longtitude']
X=mlb_data[feat]
y=mlb_data.Price
```

```python
from sklearn.model_selection import train_test_split
tr_X,v_x,tr_y,v_y=train_test_split(X,y,random_state=0)
from sklearn.tree import DecisionTreeRegressor
mlb_model=DecisionTreeRegressor(random_state=1)
mlb_model.fit(tr_X,tr_y)
```

```python
y_pred=mlb_model.predict(v_x)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(v_y,y_pred)
```

```python
def get_mae(max_leaf_nodes,tr_X,v_x,tr_y,v_y):
  model=DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state=0)
  model.fit(tr_X,tr_y)
  y_pred=model.predict(v_x)
  mae=mean_absolute_error(v_y,y_pred)
  return mae
```

```python
for max_nodes_leaf in [5,50,60,75]:
  my_mae=get_mae(max_nodes_leaf,tr_X,v_x,tr_y,v_y)
  print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_nodes_leaf,my_mae))
```
